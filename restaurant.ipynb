```python
import streamlit as st
import pandas as pd

# قاعدة بيانات بسيطة
restaurants = pd.DataFrame({
    "Nom": ["Pizza House", "Sushi Time", "Tajine Marocain", "Burger King"],
    "Ville": ["Rabat", "Casablanca", "Rabat", "Marrakech"],
    "Prix": [80, 150, 60, 70],
    "Type": ["Pizza", "Sushi", "Marocain", "Burger"]
})

st.title("Recherche de Restaurants")

ville = st.text_input("دخل المدينة")

if ville:
    resultats = restaurants[
        restaurants["Ville"].str.lower() == ville.lower()
    ]

    if len(resultats) > 0:
        st.dataframe(resultats)
    else:
        st.write("ما لقيتش مطاعم فهاد المدينة")
```
